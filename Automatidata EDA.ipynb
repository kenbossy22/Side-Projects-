```python
## Graph a Box Plot of Trip Duration and Trip Distance to check for outliers 


# Calculate trip duration in minutes
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Create a figure and axis objects
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))

# Box plot for trip_duration using seaborn (with time in minutes)
sns.boxplot(data=df, y='trip_duration', ax=axes[0])
axes[0].set_title('Trip Duration')
axes[0].set_ylabel('Duration (minutes)')

# Box plot for trip_distance using seaborn
sns.boxplot(data=df, y='trip_distance', ax=axes[1])
axes[1].set_title('Trip Distance')
axes[1].set_ylabel('Distance (miles)')

# Adjust the spacing between subplots
fig.tight_layout()

# Display the plot
plt.show()
```

```python
# Assuming you have a DataFrame named "df" with the first column as the user_id and 'trip_duration' column

# Label the first column as 'user_id'
df.columns = ['user_id'] + list(df.columns[1:])

# Filter trips with duration > 300 minutes
long_trips = df[df['trip_duration'] > 300]

# Select 'user_id' and 'trip_duration' for long trips
long_trips_info = long_trips[['user_id', 'trip_duration']]

# Display the user_id and trip_duration for long trips
print(long_trips_info)

```

## It seems as though the trip durations that exceed 300 minutes are clerical errors that mixed up the pickup and drop off times. They listed the drop off as the pickup time and vise versa. We should exclude these results, as we can't be 100% sure we can simply swap them. We should also look to remove the negative values in the Trip Duration for the by the same logic. 

```python
# Filter out trip durations less than or equal to 300 minutes and greater than 0 minutes
filtered_trips = df[(df['trip_duration'] < 300) & (df['trip_duration'] > 0)]

# Display the filtered user_id and trip_duration for long trips
filtered_trips_stats = filtered_trips.describe()
print(filtered_trips_stats)

# Create a box plot for trip_duration using seaborn
sns.boxplot(data=filtered_trips, y='trip_duration')
plt.title('Filtered Trip Duration')
plt.ylabel('Duration (minutes)')

# Display the plot
plt.show()

```

This shows us that the usual range for Trip Duration is between 2.84 and 8.53 minutes. 

```python
#Check the Total Amounts column for outliers 

# Descriptive statistics for 'total_amount'
total_amount_stats = df['total_amount'].describe()
print(total_amount_stats)

# Box plot for 'total_amount'
sns.boxplot(data=df, y='total_amount')
plt.title('Total Amount')
plt.ylabel('Amount')

# Display the plot
plt.show()
```

Here there is one value at 1200 for Total Amounts that is a clerical error, and after investigating the data in Excel, there are 14 negative values, between -3.3 and -5.8 and one value at -120. These appear to be refunds but we'll have to double-check this with the data source provider NYC Taxi and Limousine Commission. 

```python
#Filter out outlier values and plot the box plot again 

# Filter out values greater than 100 and trip_distance equals 0 to get rid out outliers 
filtered_df = df[(df['total_amount'] <= 100) & (df['trip_distance'] == 0)]

# Descriptive statistics for filtered 'total_amount'
filtered_stats = filtered_df['total_amount'].describe()
print(filtered_stats)

# Box plot for filtered 'total_amount'
sns.boxplot(data=filtered_df, y='total_amount')
plt.title('Filtered Total Amount (Trip Distance = 0)')
plt.ylabel('Amount')

# Display the plot
plt.show()
```

This shows us that the usual range of Total Amounts for each trip is between around $3.80 and $25.80 

```python


# Convert 'tpep_pickup_datetime' to datetime format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])

# Extract month and hour from the datetime column
df['month'] = df['tpep_pickup_datetime'].dt.month
df['hour'] = df['tpep_pickup_datetime'].dt.hour

# Pivot the data to create a matrix for average total amounts by month and hour
heatmap_data = df.pivot_table(index='hour', columns='month', values='total_amount', aggfunc='mean')

# Create the heatmap plot
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".2f", cbar=True)

# Set the x-axis tick labels for all 24 hours
plt.xticks(range(heatmap_data.shape[1]), [str(i) for i in range(heatmap_data.shape[1])])

plt.title('Average Total Amounts by Hour and Month')
plt.xlabel('Month')
plt.ylabel('Hour')

# Display the plot
plt.show()
```

```python
#Graph the Average Total Amounts by Month 

# Convert 'tpep_pickup_datetime' to datetime format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])

# Extract month from the datetime column
df['month'] = df['tpep_pickup_datetime'].dt.month

# Calculate average total amounts per month
avg_total_amounts = df.groupby('month')['total_amount'].mean()

# Create a line plot
plt.plot(avg_total_amounts.index, avg_total_amounts.values, marker='o')
plt.title('Average Total Amounts per Month')
plt.xlabel('Month')
plt.ylabel('Average Total Amount')

# Customize x-axis tick labels with month names
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(avg_total_amounts.index, month_names)

# Display the plot
plt.show()
```

This shows us that January, April, and July are all months that generate less revenue than others while May, June, November, and December all have higher average total amounts. This may suggest that there are more visitors to the NYC area during the early summer months and around the Thanksgiving/Christmas holiday season.
